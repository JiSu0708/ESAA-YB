앤스콤 4분할 그래프 살펴보기  
앤크콤 4분할 그래프(Anscombe's quarter): 데이터 시각화를 보여주는 전형적인 사례. 데이터를 시각화하지 않고 수치만 확인할 때 발생할 수 있는 함정을 보여주기 위해 만든 그래프.

4개의 데이터 그룹은 각각 평균, 분산과 같은 수칫값이나 상관관계, 회귀선이 같다는 특징이 있다.그래서 이런 결과만 보고 '데이터 그룹 I, II, III, IV의 데이터는 모두 같을 것이다'라고 착각할 수 있다는 것이 앤스콤이 지적한 '함정'이다.

이산형 변수- 명확하게 구분되는 값 (ex. Female/Male), 연속형 변수- 명확하게 셀 수 없는 범위의 값 (ex. Tip)

8. boxplot 메서드를 사용해 박스 그래프 그리기

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
              tips[tips['sex'] == 'Male']['tip']],
              labels=['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

**다변량 그래프 그리기**  
다변량 그래프- 3개 이상의 변수를 사용한 그래프  

**다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프**

1.앞에서 실습한 산점도 그래프(scatter_plot)에 성별을 새 변수로 추가. 성별은 색상으로 구분하면 적당.  
but, 문자열은 산점도 그래프의 색상을 지정하는 값으로 사용할 수 없으므로 0,1과 같은 정수를 사용해야 함.  
-> 문자열을 정수로 치환하는 함수 만들기

```python
def recode_sex(sex):
  if sex=='Female':
    return 0
  else:
    return 1
```

2. recode_sex가 반환한 값(0 or 1)을 데이터프레임에 추가  
sex_color 열을 새로 생성하여 추가. sex열에 recode_sex 함수를 브로드캐스팅하기 위하여 apply 메서드를 사용함.

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

3. 테이블당 인원 수(size) 산점도 그래프에 추가. (점의 크기로 표현하면 적당)  
scatter 메서드에 s, c 인자값으로 테이블당 인원 수와 성별의 치환값을 전달.

```python
# s = size(점의 크기), c = color(점의 색상)
# alpha 인자값을 0.5로 지정하여 점의 투명도를 보기 좋게 조절함.
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,
    c=tips['sex_color'],
    alpha=0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

seaborn 라이브러리를 활용하면 matplotlib보다 좀 더 화려한 그래프를 그릴 수 있다. (seaborn 라이브러리는 matplotlib 라이브러리를 기반으로 만든 라이브러리)

**단변량 그래프 그리기 - 히스토그램**

1. seaborn 라이브러리를 sns라는 이름으로 줄여 사용. tips 데이터 집합 불러오기.

```python
import seaborn as sns

tips = sns.load_dataset("tips")
```

2. subplots, distplot 메서드를 사용해 seaborn 라이브러리로 히스토그램 그리기.  
subplots 메서드로 기본 틀을 만들고 -> distplot 메서드에 total_bill 열 데이터 전달

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

3. displot 메서드를 사용하면 히스토그램과 밀집도 그래프를 같이 그린다. 밀집도 그래프를 제거하고 싶다면 kde 인자값을 False로 설정.

4. 밀집도 그래프만 나타내려면 hist 인자를 False로 지정

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

5. 여러 데이터 한 번에 출력 - 실무 환경에서 데이터 시각화를 더 효율적으로 하기 위해  
다음은 히스토그램과 밀집도를 그리는 distpot 메서드에 rug 인자를 추가하여 양탄자(rug) 그래프까지 그린 것.  
(양탄자 그래프: 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도를 표현한 그래프)

**다양한 종류의 이변량 그래프 그리기**

1. seaborn 라이브러리로 산점도 그래프 그리기  
- seaborn 라이브러리는 matplotlib 라이브러리보다 다양한 방법으로 산점도 그래프를 그릴 수 있다.  
- seaborn 라이브러리의 regplot 메서드 이용  
- regplot 메서드 이용 -> 산점도 그래프, 회귀선 함께 그려짐 -> 회귀선을 제거하려면 fit-reg = False로 지정


```python
ax = plt.subplots( )
ax = sns.regplot(x='total_bill', y='tip', data=tips) 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')     
```

```python
ax = plt.subplots( )
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False) 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

2. jointplot: 산점도 그래프&히스토그램 한 번에 그려주는 메서드  
x,y 인자- 원하는 열 이름, data 인자값- 데이터프레임 지정

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips) 
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip') 
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

3. 산점도 그래프- 점이 겹쳐 보일 경우 구분하기 어렵다는 단점
-> 육각 그래프(hexbin): 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프. 특정 데이터의개수가 많아지면 점점 진한 색으로 표현됨.  
kind = hex로 지정   
-> 육각 그래프 사용하면 데이터를 구분하기 쉽게 그린다.

```python
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

4. 이차원 밀집도 그리기  
kdeplot 메서드로 만듦. shade = True로 지정하면 그래프에 음영 효과를 줄 수 있음.

```python
ax=plt.subplots()
ax=sns.kdeplot(x=tips['total_bill'], y=tips['tip'], shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

9. 관계 그래프를 중복된 정보가 표현된다는 단점이 있다. 대각선을 기준으로 절반씩 같음.  
-> 중복된 그래프가 그려지는 위치를 직접 지정하여 원하는 그래프로 교체.  
-> map_upper, map_lower, map_diag

```python
pair_grid = sns.PairGrid(tips) 
pair_grid = pair_grid.map_upper(sns.regplot)  # 산점도 그래프
pair_grid = pair_grid.map_lower(sns.kdeplot)  # 이차원 밀집도
pair_grid = pair_grid.map_diag(sns.distplot, rug=True) # 히스토그램, 밀집도 
plt.show()
```

**04-5. seaborn 라이브러리로 그래프 스타일 설정하기**  
- set_style 메서드 사용  
스타일을 적용할 부분에서 한 번만 실행하면 이후의 그래프 모두 지정한 스타일로 바뀜.  
seaborn 라이브러리 스타일(5종류): darkgrid, whitegrid, dark, white, ticks

그래프에 스타일 지정하기

```python
fig, ax = plt.subplots( )
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

whitegrid- 그래프 배경에 가로 줄이 생긴다

```python
sns.set_style('whitegrid') 
fig, ax = plt.subplots( )
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

for문 이용해 모든 스타일 하나씩 적용해보기

```python
fig = plt.figure( )
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
for idx, style in enumerate(seaborn_styles):
  plot_position = idx + 1
  with sns.axes_style(style):
    ax = fig.add_subplot(2, 3, plot_position)
    violin = sns.violinplot(x='time' ,y='total_bill', data=tips,ax=ax)
    violin.set_title(style)
fig.tight_layout()
```
